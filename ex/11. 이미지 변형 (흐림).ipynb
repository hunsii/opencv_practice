# 11. 이미지 변형 (흐림)

## 가우시안 블러

### 커널 사이즈 변화에 따른 흐림

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('img.jpg')

# (3, 3), (5, 5), (7, 7)을 주로 사용
kernel_3 = cv2.GaussianBlur(img, (3, 3), 0) # 0은 자동으로 표준편차를 설정
kernel_5 = cv2.GaussianBlur(img, (5, 5), 0) # 0은 자동으로 표준편차를 설정
kernel_7 = cv2.GaussianBlur(img, (7, 7), 0) # 0은 자동으로 표준편차를 설정

plt.figure(figsize=(16,8))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(kernel_3, cv2.cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(kernel_5, cv2.cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(kernel_7, cv2.cv2.COLOR_BGR2RGB))
```

### 표준 편차 변화에 따른 흐림

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('img.jpg')

# (3, 3), (5, 5), (7, 7)을 주로 사용
sigma_1 = cv2.GaussianBlur(img, (0, 0), 1) # sigmaX - 가우시안 커널의 x 방향의 표준편차
sigma_2 = cv2.GaussianBlur(img, (0, 0), 2) # 
sigma_3 = cv2.GaussianBlur(img, (0, 0), 3) # 

plt.figure(figsize=(16,8))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(sigma_1, cv2.cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(sigma_2, cv2.cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(sigma_3, cv2.cv2.COLOR_BGR2RGB))
```
