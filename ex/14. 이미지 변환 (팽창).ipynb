# 14. 이미지 변환 (팽창)

## 이미지를 확장하여 작은 구멍을 채움
흰색 영역의 외곽 픽셀 주변에 흰색을 추가

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

kernel = np.ones((3, 3), dtype = np.uint8)
# print(kernel)

img = cv2.imread('dilate.png', cv2.IMREAD_GRAYSCALE)
dilate1 = cv2.dilate(img, kernel, iterations = 1) # 반복 횟수
dilate2 = cv2.dilate(img, kernel, iterations = 2) # 반복 횟수
dilate3 = cv2.dilate(img, kernel, iterations = 3) # 반복 횟수

plt.figure(figsize = (32, 16))
plt.subplot(4, 1, 1)
plt.imshow(img, 'gray')
plt.subplot(4, 1, 2)
plt.imshow(dilate1, 'gray')
plt.subplot(4, 1, 3)
plt.imshow(dilate2, 'gray')
plt.subplot(4, 1, 4)
plt.imshow(dilate3, 'gray')
```
