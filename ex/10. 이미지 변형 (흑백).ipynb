# 10. 이미지 변형 (흑백)

이미지를 흑백으로 읽음

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, 'gray')
```

불러온 이미지를 흑백으로 변경

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('img.jpg')

dst = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(dst, 'gray')
```
